```python
m1 = pd.DataFrame({'A' : ['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                   index = [0,1,2,3])
```

```python
m2 = pd.DataFrame({'A' : ['A0','A8','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                   index = [0,1,2,3])
```

```python
m1
```

```python
m2
```

```python
pd.merge(m1,m2)  #we dont have A5 in both so skipped
```

```python
pd.merge(m1,m2, on = 'A')
```

```python
# INNER MEANS COMMON DATA SET, INTERSECTION
pd.merge(m1,m2, on = 'A', how = 'inner')
```

```python
pd.merge(m1,m2, on = 'A', how = 'left') 
# extract all the data from left side and
# if same data is not found in the right side fill it with null
```

```python
pd.merge(m1,m2, how = 'right')
```

```python
pd.merge(m1,m2, on ='A', how = 'right')
```

```python
pd.merge(m1,m2, on =['A','B'], how = 'right')
```
