```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips);
```

```python
# create plot between 2 things and then categorised it by smoker or non smoker
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker');
# concluison - person who has given the highest tips is a smoker
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'time');
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', style = 'smoker');
# hue seperate the data by color and style seprate the data by shape
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = tips, style = 'size', hue = 'size');
```
