 ### scatter plot 
 * generating graph bw x,y and the color of the dot will be decided by c range
 * for example 1 row x = 5.1 y =1.4 (yaha se point locate kiye)
 * now color of point decidecd by c and c ranges in between 2 to 4 in our data 

```python
iris.plot.scatter(x = 'sepal_length', y = 'petal_length', figsize = (5,5), c = 'sepal_width', s = 100);
# generating graph bw x,y and the color of the dot will be decided by c range
# for example 1 row x = 5.1 y =1.4 (yaha se point locate kiye)
# now color of point decidecd by c and c ranges in between 2 to 4 in our data 
```
